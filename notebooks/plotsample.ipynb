```python
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
plt.xkcd()
plt.figure(figsize=(12,10))

x = np.linspace(0, 5, 10)
y = x ** 2

x2 = np.linspace(-4, 4, 1024)
y2 = .25 * (x2 + 4.) * (x2 + 1.) * (x2 - 2.)

plt.subplot(2,3,1)
plt.plot(x, y, 'r--')
plt.subplot(2,3,2)
plt.plot(y, x, 'g*-')
plt.subplot(2,3,3)
plt.plot(x, y, 'b--')
plt.subplot(2,3,4)
plt.plot(y, x, 'y*-')

plt.subplot(2,3,5)
plt.plot(x2, y2, 'b--')
plt.subplot(2,3,6)
plt.plot(y2, x2, 'y*-')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N))**2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
